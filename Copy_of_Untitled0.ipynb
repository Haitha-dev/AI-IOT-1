```python
#step-3(handling missing data or values)
#>senario-1:- missing values in numerical columns
#sol:-caluclate mean or Average of the coloum>>we have to write in num.py
#>senario-2:-Missing Values in other than numerical values in column
#sol- fill the with high repeated value.
#>senario-3:-if we have missing values in huge dataset(eg:-20000 records in that 2 rows have empyt values)
#sol:-remove 2those empty values from the dataset.
import numpy as np
from sklearn.impute import SimpleImputer
#sklearn is big library
# simpleimpute(it will use in the fill the blanks in the data) is class  functions which is from sklearn.impute   so we import it
#sklearn is also called as scikit is python libery widely used for machine learning and data moduling
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')#created impute ,ean methods and giving input
imputer.fit(X[:,1:3])#applying prepare for columsby using .fit
X[:,1:3]=imputer.transform(X[:,1:3])#performing actions using transfefuncton replaying insted of nan value to mean value
```

```python
X
```

```python
#step4:-handling the catagorial data(insted of number there is  text formate is called catagorile data)
#coverting catagorial data to numerical data
from sklearn.preprocessing import OneHotEncoder#it was a librabry in sklearn it is label encoder it is covering text into  numerics
from sklearn.compose import ColumnTransformer# colum transformer is a class imported from sklearn.compose>it is used to apply transeformes or changes to colums
# columTransformer is class which is used to apply the changes of numerical to text
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')#encoder is a method converts text to numerics
X=ct.fit_transform(X)
#OneHotEncoder is only upadte for input columns
#linearencoder is only update for the output colum
```
